```python
#bai 11

a,b,c = map(int,input().split())
edge = [a, b, c]
sum = a+b+c
isTriangle = True

for i in edge:
    if i>=sum-i:
        isTriangle = False
        break

if isTriangle:
    if a==b==c :
      print("Tam giac deu")
    else :
      if a==b or b==c or c==a :
        print("Tam gia can")
      else :
        print("Tam giac thuong")
else:
    print("Khong phai tam giac")
```

```python
#bai 12
def bac(so_dien):
    if 0<=so_dien<=50:
        return 1500
    if 51<=so_dien<=100:
        return 2000
    if so_dien>100:
        return 3000

so_dien = int(input())
tien = 0
while (so_dien!=0):
    tien += bac(so_dien)
    so_dien -= 1

print(tien)
```
